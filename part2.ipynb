```python
import pandas as pd
import sympy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df[df["Name"]=="Jolteon"]
```

```python
df_type=df[(df["Type 1"]=="Grass") | (df["Type 1"]=="Electric") | (df["Type 1"]=="Dragon") | (df["Type 1"]=="Ground")]
```

```python
df_type_gen1=df_type[df_type["Generation"]==1]
```

```python
df_type_gen1_sorted=df_type_gen1.sort_values("Sp. Def")
display(df_type_gen1_sorted)
```

```python
sns.boxplot(x="Generation",y="HP",data=df)
```

```python
quantile_97=df["HP"].quantile(0.97)
display(quantile_97)
```

```python
hp_outliers_df=df[df["HP"]>quantile_97]
```

```python
hp_outliers_id = hp_outliers_df.index.tolist()
```

```python
df_hp_no_outliers=df.drop(hp_outliers_id)
```
